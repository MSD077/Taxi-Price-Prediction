```python
from pyspark import SparkContext
from pyspark.sql import SQLContext
from pyspark.sql.functions import *
sc = SparkContext.getOrCreate();
spark = SQLContext(sc)
from pyspark.ml.feature import StringIndexer
from pyspark.ml.regression import RandomForestRegressor
from pyspark.ml import Pipeline
from pyspark.ml.feature import VectorAssembler
from pyspark.ml.evaluation import RegressionEvaluator
from pyspark.ml.stat import Correlation
```

```python
path = "/home/sasidev/pyspark_practice/Chicago_Taxi.csv"
```

```python
df = spark.read.format('com.databricks.spark.csv').options(header='true', inferschema='true').load(path)
df.show()
```

```python
dict = {row['payment_type']:row['trip_total'] for row in df.collect()}
print(dict)
```

```python
df.toPandas().info()
```

```python
df.toPandas().info()
```

```python
indexer = StringIndexer(inputCol="payment_type",outputCol = "payment_type_indexed")
df = indexer.fit(df).transform(df)
```

```python
df = df.drop('trip_start_timestamp','trip_end_timestamp','company','payment_type')
```

```python
df.toPandas().info()
```

```python
assembler = VectorAssembler(
    inputCols=["trip_seconds", "trip_miles", 
               "pickup_community_area",'dropoff_community_area','payment_type_indexed'],
    outputCol="features")
```

```python
output = assembler.transform(df)
```

```python
final_df = output.select("features",'trip_total')
```

```python
train, test = final_df.randomSplit([0.75, 0.25], seed = 42)
```

```python
rf = RandomForestRegressor(labelCol='trip_total')
```

```python
model = rf.fit(train)
```

```python
pred = model.transform(test)
```

```python
pred.show()
```

```python
pred.select("prediction","trip_total","features")
```

```python
evaluator = RegressionEvaluator(labelCol="trip_total", predictionCol="prediction", metricName="rmse")
```

```python
rmse = evaluator.evaluate(pred)
```
